```python
from pathlib import Path
import json
import pandas as pd
```

```python
METHOD_DIR = "../../ai/data/pmc_txt/method"
RESULT_DIR = "../../ai/data/pmc_txt/result"
```

```python
method_dir_path = Path(METHOD_DIR)
result_dir_path = Path(RESULT_DIR)
```

```python
method_file_path_list = list(method_dir_path.glob("*.json"))
result_file_path_list = list(result_dir_path.glob("*.json"))
```

```python
method_file_path_list
```

```python
# Check if index 0s are all header title.
for method_file_path in method_file_path_list:
    with open(method_file_path) as f:
        method_text_list = json.load(f)
    print(method_text_list[0])
```

```python
method_file_path_list[0].stem
```

```python
method_concat_list = []
method_pmid_list = []
for method_file_path in method_file_path_list:
    with open(method_file_path) as f:
        method_text_list = json.load(f)[1:]
    method_text = " ".join(method_text_list)
    method_concat_list.append(method_text)
    method_pmid_list.append(method_file_path.stem)
```

```python
result_concat_list = []
result_pmid_list = []
for result_file_path in result_file_path_list:
    with open(result_file_path) as f:
        result_text_list = json.load(f)[1:]
    result_text = " ".join(result_text_list)
    result_concat_list.append(result_text)
    result_pmid_list.append(result_file_path.stem)
```

```python
method_concat_list
```

```python
result_concat_list
```

```python
method_dict = {
    "pmid": method_pmid_list,
    "text": method_concat_list
}

result_dict = {
    "pmid": result_pmid_list,
    "text": result_concat_list
}
```

```python
method_df = pd.DataFrame.from_dict(method_dict)
result_df = pd.DataFrame.from_dict(result_dict)
```

```python
method_df.to_csv("../../ai/data/pmc_txt_concat/method_concat.csv")
result_df.to_csv("../../ai/data/pmc_txt_concat/result_concat.csv")
```
